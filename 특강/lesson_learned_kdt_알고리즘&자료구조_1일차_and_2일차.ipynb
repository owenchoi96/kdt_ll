```python
# delete 함수. 
def delete_data(position):
    katok[position] = None
    kLen = len(katok)
    
    for i in range(position+1, kLen, 1):
        katok[i-1] = katok[i]
        katok[i] = None
    
    del katok[kLen-1] # 마지막 칸 삭제

delete_data(4)
print(katok)
```

## 단순 연결 리스트

```python
# 단순 연결 리스트란?
# : 떨어진 곳에 위치한 데이터를 화살표로 연결한 것. 

# 단순 연결 리스트의 개념
# : 노드들이 물리적으로 떨어진 곳에 위치
# : 각 노드의 번지도 순차적이지 않음
# : 화살표로 표시된 연결(링크,link)을 따라가면 선형 리스트 순서와 같음. 

# 사용자 입장에서는 선형 리스트로 했든 연결 리스트로 했는지 모름
# 하지만 프로그래머 입장에서 둘은 다름. 
# 선형 리스트는 물리적으로 붙어있음
# 연결 리스트는 물리적으로 떨어져 있지만, 논리적으로 연결되어있음. 

# 데이터와 화살표를 합쳐서 '노드'라고 부름. 

# 연결 리스트로 데이터를 삽입/삭제 할때
    # 선형 리스트는 많은 작업이 필요 (오버헤드 발생) -> 안되는 건 오버헤드가 아니지만
    #                                         무리하게 발생함. 
    # 단순 연결 리스트는 해당 노드의 앞뒤 링크만 수정하면 되므로 오버헤드가 거의 발생하지 않음. 
    
# 단순 연결 리스트의 원리
# 노드 구조: 
    # 단순 연결 리스트는 다음 데이터를 가리키는 링크가 더 필요
    # 노드는 데이터와 링크로 구성된 항목
    # head : 연결 리스트의 가정 첫번째를 잡는 노드. 
    # 나머지는 연결되어 있기 때문에 head만 잘 잡고 있으면 접근이 가능함. 
    
    
# 구현
# 연결 리스트 만드는 방법. 

## 함수
class Node():
    
    def __init__(self):
        self.data = None
        self.link = None
        
node1 = Node() # 텅빈 노드가 만들어짐
node1.data = '다현' # 데이터 부분에 다현을 넣음 

node2 = Node()
node2.data = '정현'
node1.link = node2 # node1의 링크를 node2

node3 = Node()
node3.data = '쯔위'
node2.link = node3

node4 = Node()
node4.data = '사나'
node3.link = node4

node5 = Node()
node5.data = '지효'
node4.link = node5

## 노드의 삽입
# newNode = Node()
# newNode.data = '재남'
# newNode.link = node2.link # 재남도 쯔위를 가리키고 정연도 쯔위를 가리킴
# node2.link = newNode # 정연의 링크를 재남으로 변경. 

## 노드의 삭제
# 쯔위의 탈퇴
node2.link = node3.link # 쯔위를 건너뛴 다음 link로 선언해주고
del(node3) # 쯔위 삭제 


# 첫번째 노드를 통해서 .link로 다른 노드에 접근할 수 있음. 
print(node1.data, end=' ')
print(node1.link.data, end=' ') # print(node2.data)
print(node1.link.link.data, end=' ') # print(node3.data)
print(node1.link.link.link.data, end=' ') # print(node4.data)
```

```python
#### 실무 버전의 연결 리스트 ####
# 메모리 : 연결 리스트가 전부 들어있음. -> 메모리
# head : 연결 리스트의 첫번째 부분을 가리키는 노드
# current : 현재 작업 중인 노드
# pre : 현재 처리 중인 노드의 바로 앞 노드 

# 처음에는 모두 비어있고 다음과 같이 초기화
# memory = []
# head, current, pre = None, None, None

## 함수 #####################################################################################
class Node():
    
    def __init__(self):
        self.data = None
        self.link = None
        
def printNodes(start) : 
    current = start
    print(current.data, end=' ')

    while current.link != None:
        current = current.link
        print(current.data, end=' ')
    print()

### 삽입 함수 ###
def insertNode(findData, insertData):
    global memory, head, pre, current
    # case 1: head 앞에 삽입할 때 (다현, 화사)
    if (findData == head.data):
        node = Node()
        node.data = insertData
        node.link = head
        head = node
        memory.append(node)
        return # 더 이상 할게 없으면 return
    
    # case 2: 중간 노드에 삽입 할 때 (사나, 솔라)
    current = head
    while (current.link != None):
        pre = current # 다음으로 가기 전에 pre한테 맡겨두고 감. 
        current = current.link # 다음으로 가는 방법 
        if (current.data == findData):
            node = Node()
            node.data = insertData
            node.link = current 
            pre.link = node # pre 에다가 방금 만든 node를 넣어줌
            return # 더 이상 할게 없다면 return. 
        
    # 여기까지 왔다는 뜻은 첫번째는 case 1에서 걸리지 않고 case 2에서 걸리지 않았기 때문에
    # 그냥 마지막에 insertData를 넣어주기. 
    # case 3: 없는 노드 앞에 삽입 할때
    node = Node()
    node.data = insertData
    current.link = node
    memory.append(node)
    return

### 삭제 함수 ###
def deleteNode(delData):
    global memory, head, pre, current
    
    # Case 1 : 삭제할 데이터가 head인 경우
    if (delData == head.data):
        current = head
        head = head.link # 다음 것을 head로 지정. head.link는 head 다음의 노드
        del(current)
        return # 작업끝
    
    # Case 2 : 삭제할 데이터가 중간인 이유
    # 여기서는 머리가 아님, 쯔위를 찾기
    current = head # 처음부터 찾기
    while (current.link != None): # 다음 링크가 없을 때까지 (끝까지)
        pre = current # pre한테 current를 맡김
        current = current.link # current는 다음으로 감
        if (current.data == delData):
            pre.link = current.link
            del(current)
            return
        
    # Case 3 : 없는 데이터를 삭제할 경우 (재남)
    return # 원래는 할거 없음 
           # 케이스를 꼭 나눈다면 return만 쓰면 됨. 
    
### 검색 노드 ###
def findNode(findData):
    global memory, head, pre, current
    
    # case 1 : 찾고자 하는 노드가 첫번째 에 있을 때
    current = head 
    if (current.data == findData):
        return current # current의 data가 아니라 노드 통째로 꺼내야 함. 
    
    # case 2 : 처음에 없을 때
    while (current.link != None):
        current = current.link # 다음 노드로 넘어가기. 
        if (current.data == findData):
            return current 
    
    return Node() # 끝까지 못 찾으면 비어있는 노드를 리턴


## 전역변수 #####################################################################################
memory = []
head, current, pre = None, None, None
dataAry = [
    '다현', '정연', '쯔위', '사나', '지효'
] 

## main #####################################################################################

# 하나짜리 연결리스트 완성. 
node = Node()
node.data = dataAry[0]
head = node
memory.append(node)

for data in dataAry[1:]: # dataAry의 첫번째를 제외한 리스트 안의 다음 값부터...
    pre = node  # 노드를 재활용하기 전에 pre에 이전(첫번째) node를 저장
    node = Node()
    node.data = data
    pre.link = node  # 다음의 첫번째 노드의 link에 다음 노드를 넣어줌
    memory.append(node)

# insertNode cases
# insertNode('다현', '화사') 다현이 첫번째 노드에 있을 때
# insertNode('사나', '솔라')
# insertNode('재남', '문별') # 재남이 현재 없음 -> 문별을 마지막 노드에 삽입해줌. 

# deleteNodes cases
# deleteNode('다현') 처음 데이터를 삭제할 경우
# deleteNode('쯔위') 중간에 데이터가 있는 경우
# deleteNode('재남') 데이터가 없는 경우

# printNodes(head) # 함수를 이용하여 리스트 안에 데이터 다 뽑아오기. 

fNode = findNode('사나')
print(fNode.data, '뮤비가 나옵니다. 큥짝~~')
```

## 스택

```python
# 스택 구조란? 
# ex) 아이스크림 콘에 여러 가지 맛을 쌓을 때, 가장 먼저 넣은 맛을 가장 나중에 먹을 수 있는 구조

# 스택의 개념
    # 스택(stack) 자료구조는 한쪽 끝이 막힌 형태
    # 입구가 하나이기 때문에 먼저 들어간 것이 가장 나중에 나오는 구조(선입후출, 후입선출) 
    # FILO (First In Last Out)
    
# 스택 원리
# 스택 기본 구조
    # 스택에 데이터를 삽입하는 작동 : push
    # 스택에 데이터를 추출하는 작동 : pop
    # 스택에 들어 있는 가장 위의 데이터 : top
    
# 데이터 삽입 : push
    # 데이터 초깃값을 -1로 지정 -> 다른 숫자로 하면 다른 데이터가 있는 줄로 착각할 수 있음.
# 데이터 추출 : pop
    # top 위치의 데이터를 추출 (pop)
    # top을 한 칸 아래로 이동 
    # top의 값이 -1이면 아무것도 없음을 의미. 
```

```python
## 함수

## 전역 변수
SIZE = 5
stack = [None for _ in range(SIZE)] # 리스트 컴프리헨션
top = -1

## 메인
# Push()
top += 1
stack[top] = '커피'

top += 1
stack[top] = '녹차'

top += 1
stack[top] = '꿀물'

print('바닥: ', stack)

# Pop() : 추출 (삭제보다는 추출에 더 가까운 의미)
data = stack[top]
stack[top] = None
top -= 1
print('Pop --> ', data)

data = stack[top]
stack[top] = None
top -= 1
print('Pop --> ', data)

data = stack[top]
stack[top] = None
top -= 1
print('Pop --> ', data)
print('바닥: ', stack)
```

```python
## 실무버전 스택 만들기 ##
## 데이터 삽입 과정 ##
# 스택이 꽉 찼는지 확인하는 함수
    # top 값이 '스택 크기-1'과 같다면 스택이 꽉 찬 상태
    # 꽉 찬 상태 -> 오버플로어 : 굉장히 좋지 않음. 


## 함수
def isStackFull():
    # 지역변수가 아닌 global 함수로 인식하겠다
    global SIZE, stack, top # 스택의 최상단 데이터를 가리키는 변수
    if (top == SIZE-1): # 탑이 꼭대기를 가리키면
        return True
    else:
        return False

def push(data):
    global SIZE, stack, top
    if (isStackFull()):
        print('스택 꽉!')
        return
    top += 1
    stack[top] = data
    
def isStackEmpty():
    global SIZE, stack, top
    if (top == -1):
        return True
    else:
        return False

    
def pop():
    global SIZE, stack, top
    # 텅빈 상태에서 pop하면 언더플로어
    if (isStackEmpty()):
        print('스택이 텅 비었습니다')
        return
    data = stack[top]
    stack[top] = None
    top -= 1
    return data 

def peek(): # 다음에 나올 데이터를 미리 확인하는 함수 (실제 뽑는 것은 아님.)
    global SIZE, stack, top
    if (isStackEmpty()):
        print('스택이 텅 비었습니다')
        return
    return stack[top]
    


## 전역 변수
SIZE = 5
stack = [None for _ in range(SIZE)] # 리스트 컴프리헨션
top = -1

## 메인
push('커피')
push('녹차')
push('꿀물')
push('콜라')
push('환타')
# print(stack)

retData = pop()
print('팝 --> ', retData)

retData = pop()
print('팝 --> ', retData)

retData = pop()
print('팝 --> ', retData)
print(stack)
```

**순차큐**

```python
## 함수
# 큐가 꽉 찼는지 먼저 확인하기
# ----------- 코드 수정 ---------------
# 꽉 찾는지 확인할 때 3가지 케이스
def isQueueuFull():
    global SIZE, queue, front, rear
    
    if (rear != SIZE-1): # 뒤에 공간이 있을 때
        return False
    elif (rear == SIZE-1) & (front==-1): # 뒤에는 공간이 없는데
        return True                      # 앞에는 비어 있을 때
    elif (rear == SIZE-1) & (front!=-1): 
        # 한칸씩만 땡기기
        for i in range(front+1, SIZE, 1):
            queue[i-1] = queue[i]
            queue[i] = None
        front -=1 # front의 정의는 맨 앞의 친구의 앞에
        rear -=1 # rear도 하나 땡겨줘야 함. 
        return False # 꽉 안참
    


def enQueue(data):
    global SIZE, queue, front, rear
    
    if isQueueuFull():
        print('큐가 꽉 찼습니다')
        return
    rear += 1   # 손님은 뒤쪽 출구에서 들어옴. 
    queue[rear] = data
    
def isQueueEmpty():
    global SIZE, queue, front, rear
    if (front == rear):
        return True
    else:
        return False
    
    
def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 텅 비었습니다.')
        return
    front += 1
    data = queue[front]
    queue[front] = None
    return data

# 다음 대기 손님 함수 
def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 텅 비었습니다.')
        return
    return queue[front+1]



## 전역
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1


## 메인

enQueue('화사')
enQueue('솔라')
enQueue('문별')
enQueue('휘인')
enQueue('선미')
# enQueue('재남')

print('출구<--', queue, '<--입구')

retData = deQueue()
nextData = peek()
print('*식사 손님 이쪽으로: ', retData)
print('-----> 다음 대기 손님: ', nextData)
print('출구<--', queue, '<--입구')

enQueue('재남')
print('--------- 재남이 들어갑니다 앞으로 땡겨주세요--------')
print('출구<--', queue, '<--입구')
```

**원형 큐**

```python
# : 순차 큐를 구부려서 끝을 이은 원형 큐
# 오버헤드가 발생하지 않음. 
# 실무에서 데이터 양이 적다면 --> 순차 큐
#        데이터 양이 많다면 --> 원형 큐
# 원형 큐는 -1 자리가 없음 --> 초기값 = 0
```

```python
# 원형 큐는 앞쪽에 한칸을 못씀 
# (rear+1)%SIZE -> 일반 큐를 원형 큐로 만들어주는 마법 코드
def isQueueFull():
    global SIZE, queue, front, rear
    if ((rear+1)%SIZE == front):
        return True
    else:
        return False
    
def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print('큐가 꽉 찼습니다')
        return
    rear = (rear + 1) % SIZE # 일반 큐를 원형 큐로 바꿔주는 회심의 코드
    queue[rear] = data
    

def isQueueEmpty(): 
    global SIZE, queue, front, rear
    if (front == rear):
        return True
    else:
        return False
    
    
def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print ('큐가 비어있습니다')
        return
    front = (front+1)%SIZE # 사이즈보다 커지는 그 순간에만 0으로 바꿔줌. 
    data = queue[front]
    queue[front] = None
    return data

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 꽉 찼습니다')
        return
    return queue[(front+1)%SIZE]

## 전역
SIZE = 5
queue = [None for _ in range(SIZE)]
front=rear=0

enQueue('화사')
enQueue('솔라')
enQueue('문별')

print('출구<--', queue, '<--입구')
```

**이진 탐색 트리** (이진트리의 한 종류)

```python
# 어려움, 할만함
# 이진 트리 중 활용도가 높은 트리로, 데이터 크기를 기준으로 일정 형태로 구성함. 
# ---- 작은 거 왼쪽, 큰거 오른쪽 ---- : 단순한 규칙
# 어려운 점? : 구성이 좀 어려움
# 장점 : 여기서 무언가를 찾는게 굉장히 빠름 --> 이진 "탐색" 트리
```

**구현**

```python
## 함수
class TreeNode(): # 트리 노드의 클래스
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None



## 전역 변수
memory = [] # 저장소
root = None 
nameAry = ['블랙핑크', '레드벨벳', '마마무', '에이핑크', '걸스데이', '트와이스']


## 메인
node = TreeNode()
node.data = nameAry[0]
root = node # node1을 root로 만들기
memory.append(node)
# --- 여기까지 완벽한 하나의 이진트리 --- 

for name in nameAry[1:]: # 레드벨벳부터 나머지
    node = TreeNode()
    node.data = name
    
    # root부터 비교해서 차례로 와야됨
    current = root # current -> 현재 작업중인 노드 
    while True: # 언제 자리를 잡을 지 모르기 때문에 while=True로 무한 반복을 돌려야 함. 
                # 자리를 잡을 때까지 무한 루프를 돌려야 함. (--> 트리의 크기를 모르기 때문에)
        if (name < current.data): # 가나다 순으로
            # 왼쪽으로 내려가되 비었으면 자리를 잡지만
            # 아니라면 다시 내려가야 됨
            if (current.left == None): 
                current.left = node
                break # 자리를 잡으면 멈추기 
            # 비어있지 않으면 또 내려가. 
            current = current.left # 그리고 다시 위에 if 문으로 올라감. (while문으로 계속 반복이 됨.)
        else:
            if (current.right == None): 
                current.right = node
                break # 자리를 잡으면 멈추기
            current = current.right
        
    memory.append(node)

# print('이진 탐색 트리 구현 완료!')

# ---- 이진 탐색 트리 구현 후 이진 탐색 하는 코드 ----
findName = '바바부'
current = root 
while True: # 예측을 못하면 무한 루프를 돌려야함 
    if (findName == current.data):
        print(findName, '찾음 ~~ 뮤비 스타트')
        break
    elif (findName < current.data):
        if (current.left == None):
            print(findName, "없어요")
            break
        current = current.left
    else:
        if (current.right == None):
            print(findName, '없어요')
            break 
        current = current.right
```

```python
# 2차원 사진  배열을 좌우 반전 시킨다.
def reversalPhoto(photoAry, height, width): # 2차원 사진 배열, 사진 배열의 행 개수, 사진 배열의 열 개수
    retAry = [] # 좌우 반전된 2차원 사진 배열
    ###########   여기부터 코딩 (1) ---------------->
    
    # ----- 내가 푼것 ------
    retAry = [[None for _ in range(width)] for _ in range(height)]
    for i in range(height):
        for k in range(width):
            retAry[i][width-k-1] = photoAry[i][k]

    # ---- 선생님 코드 -----
    # retAry 똑같은 크기로 만들고 채워넣는 방향으로
    
    # 내가 푼 것이랑 동일함! 
    

    ###########   <-------------- 여기까지 코딩 (1)
    return retAry


# 2차원 사진  배열을 2배 확대 시킨다.
def zoomScale(photoAry, height, width): #2차원 사진 배열, 사진 배열의 행 개수, 사진 배열의 열 개수
    retAry = [] # 2배 확대된 2차원 사진 배열
    ###########   여기부터 코딩 (2) ---------------->
    
    retAry = [[None for _ in range(width*2)] for _ in range(height*2)]
    for i in range(height*2):
        for k in range(width*2):
            retAry[i][k] = photoAry[int(i/2)][int(k/2)]


    ###########   <-------------- 여기까지 코딩 (2)

    return retAry

# 2차원 사진 배열을 왼쪽으로 90도 회전시킨다.
def rotatePhoto(photoAry, height, width): # 2차원 사진 배열, 사진 배열의 행 개수, 사진 배열의 열 개수
    retAry = [] # 왼쪽으로 90도 회전된 2차원 사진 배열
    ###########   여기부터 코딩 (3) ---------------->
    retAry = [[None for _ in range(height)] for _ in range(width)]
    for i in range(height):
        for k in range(width):
            retAry[width-k-1][i] = photoAry[i][k]


    ###########   <-------------- 여기까지 코딩 (3)

    return retAry


# 2차원 사진 배열에 테두리를 추가한다.
def drawEdge(photoAry, height, width): # 2차원 사진 배열, 사진 배열의 행 개수, 사진 배열의 열 개수
    retAry = [] # 테두리가 추가된 2차원 사진 배열
    ###########   여기부터 코딩 (4) ---------------->

    retAry = [['*' for _ in range(width+2)] for _ in range(height+2)]
    for i in range(height):
        for k in range(width):
            retAry[i+1][k+1] = photoAry[i][k]


    ###########   <-------------- 여기까지 코딩 (4)

    return retAry


## 전역 변수 선언 부분
inputPhotoAry =[] # 입력 배열(2차원) 
outputPhotoAry = [] # 출력 배열(2차원)
rowNum, colNum=0, 0 # 배열의 크기. 

def main() :
        global inputPhotoAry, outputPhotoAry, rowNum, colNum

        loadData() # 2차원 이미지 읽어오기

        ## 원본 출력
        print(' ----- (0) 원본 사진 -----')
        printData(inputPhotoAry, rowNum, colNum)
        
        ## 좌우 반전
        outputPhotoAry = reversalPhoto(inputPhotoAry, rowNum, colNum)
        print(' ----- (1) 좌우 반전된 사진 -----')
        printData(outputPhotoAry, rowNum, colNum)
        inputPhotoAry = outputPhotoAry[:] #  출력배열 --> 입력 배열
        
        ## 2배 확대
        outputPhotoAry = zoomScale(inputPhotoAry, rowNum, colNum)
        print(' ----- (2) 2배 확대된 사진 -----')
        rowNum *= 2; # 행 2배 
        colNum *= 2;  # 열 2배
        printData(outputPhotoAry, rowNum, colNum)
        inputPhotoAry = outputPhotoAry[:] #  출력배열 --> 입력 배열

        ## 왼쪽 90도 회전
        outputPhotoAry = rotatePhoto(inputPhotoAry, rowNum, colNum)
        print(' ----- (3) 90도 회전된 사진 -----')
        rowNum, colNum = colNum, rowNum # 행, 열을 교체
        printData(outputPhotoAry, rowNum, colNum)
        inputPhotoAry = outputPhotoAry[:] #  출력배열 --> 입력 배열
        
        ## 테두리가 추가된 사진.
        outputPhotoAry = drawEdge(inputPhotoAry, rowNum, colNum)
        print(' ----- (4) 테두리가 추가된 사진 -----')
        printData(outputPhotoAry, rowNum+2, colNum+2)
        
## 함수 선언 부분
def  loadData() : # 데이터 불러오기
    global inputPhotoAry, outputPhotoAry, rowNum, colNum

    ###########
    # 제공 데이터 세트 1 
    # 4x2 사진 데이터. 
    ###########
    rowNum , colNum = 4, 2 # 행 및 열의 개수
    inputPhotoAry = \
    [
        ['C', 'D' ],
        ['K', 'P' ],
        ['A', 'R' ],
        ['P', 'Q' ]
    ]


def printData(photo, x, y) :
    global inputPhotoAry, outputPhotoAry, rowNum, colNum
    
    for i in range(0, x) :
        for k in range(0, y) :
                try :
                        print("%c " % photo[i][k], end='')
                except :
                        pass
        print()
    print()

## 메인 함수 호출 ##
if __name__ == "__main__" :
    main()

```
